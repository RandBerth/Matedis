```python
def si_solo_si(a,b):
    return a == b

def condicional(a,b):
    return not (a == True and b == False)

# casos de prueba
x = True
y = True
z = True

#(x and y) <--> z
print(si_solo_si((x and y),z))

#(x --> y) <--> no z
print(si_solo_si(condicional(x,y),(not z)))

#--------------------------------------------------------------------------------------

def nand(a,b):
    return not a or not b

# if not (x and y) then z
print(condicional(nand(x, y),z))

# (x and y) <--> not (x and y)
print(si_solo_si((x and y),nand(x,y)))
```
