### Implicación

Empezar con la implicación facilitará la explicación de la **lógica proposicional** en código.

Normalmente, expresamos la implicación o el enunciado condicional como $p$ implica $q$ con la conectiva lógica $\rightarrow$.

En código, el antecedente $p$ o la premisa sirve como argumento de la condición y a veces como parámetro, mientras que el consecuente $q$ es la acción resultante dada la condición.

Se denota formalmente como:

<h3>$$p\rightarrow q$$</h3>

También podemos visualizar esto mediante un diagrama de flujo, de modo que pueda implementarse fácilmente como un algoritmo.

if.jpeg

Como ejemplo, se nos da una declaración:

<center>
    <h4>
        <i>
"Las luces de la habitación se encenderán si el sensor detecta personas".
        </i>
    </h4>  
</center>

<center>
    Podemos entonces formalizar esto en nuestro fbf como: <br>
    $p$ = El sensor detecta personas <br>
    $q$ = Las luces se encienden <br>
    $r$ = Las luces se apagan <br>
       
</center>
<h3>$$p\rightarrow q$$</h3>

Puede codificarse como:

```python
esDetectado = False
if (esDetectado):
    print("Encender las luces.")
```

Sin embargo, cuando se habla de condiciones es natural añadir la cláusula **else** o el **consecuente negativo**.

El enunciado **else** puede formularse como: si no $q$ entonces $r$.

Aunque la reacción contraria se observará como el **consecuente alternativo** de $\neg p$ no es lo mismo que la negación de $\neg q$, por lo que se traduce como un suceso diferente $r$.

Puede entonces tener el fbf como:   

<h3>$$(p\rightarrow q)  \land (\neg p \rightarrow r) $$</h3>

Y por su diagrama de flujo:

image.png

También podemos modificar la afirmación anterior como:

<center>
    <h4>
        <i>
"Las luces de la habitación se encenderán si el sensor detecta personas, de lo contrario las luces se apagarán."
        </i>
    </h4>  
</center>

<center>
    Podemos entonces formalizar esto en nuestro fbf como: <br>
    $p$ = El sensor detecta personas <br>
    $q$ = Las luces se encienden <br>
    $r$ = Las luces se apagan <br>
       
</center>
<h3>$$(p\rightarrow q)  \land (\neg p \rightarrow r) $$</h3>

Podemos entonces codificar esto como:

```python
if (esDetectado):
    print("Encender las luces.")
else:
    print("Apagar las luces.")
```

El tercer caso de implicaciones en programación es la aparición de una tercera o enésima condición.

En otros lenguajes de programación esto puede ser en forma de sentencias **case switch** o **else if**, pero en Python se implementa como **elif**.

La sentencia **elif** o cualquier otra sentencia adicional puede intercalarse en la sentencia **if..else** como una implicación anidada $a\rightarrow b$ y su consecuente alternativo.

Podemos entonces reestructurar la fbf como:

<h3>$$(p\rightarrow q) \land  (\neg p \rightarrow ((a \rightarrow b) \land (\neg a \rightarrow b))) $$</h3>

Visualización de la fbf:

elif.jpeg

Podemos entonces intentar aumentar el número de casos de la afirmación anterior:

<center>
    <h4>
        <i>
"Las luces de la habitación se encenderán ligeramente si el sensor detecta a una persona, si hay más de una persona se encienden completamente, si no hay gente las luces se apagarán".
        </i>
    </h4>  
</center>

<center>
    Podemos entonces formalizar esto en nuestro fbf como: <br>
    $p$ = El sensor detecta a una persona <br>
    $a$ = El sensor detecta a más de una persona<br>
    $q$ = Las luces se encienden ligeramente<br>
    $b$ = Las luces se encienden completamente<br>
    $r$ = Las luces se apagan <br>
 </center>   
    <h3>$$(p\rightarrow q)\land(\neg p\rightarrow ((a \rightarrow b) \land (\neg a \rightarrow b)))$$</h3>

<h3>$$$$</h3>

Entonces podemos codificarlo como:

```python
valor_deteccion = 1
if (valor_deteccion == 0):
    print("Encendido de las luces (ligeramente).")
elif (valor_deteccion == 1):
    print("Encendido de las luces (¡al máximo!).")
else:
    print("Apagado de las luces.")
```

## Disyuncion

En contraste con el lenguaje natural y conversacional que asume que la conectiva lógica "o" es Exclusiva, la lógica de programación trata la conectiva lógica "o" como una operación de **Disyunción Inclusiva**.

Las operaciones de disyunción requieren al menos dos argumentos o premisas, podemos denotarlas como $p$ y $q$ de nuevo y usamos la conectiva lógica $\lor$.

Podríamos entonces escribir la fbf como:

<h3>$$p\lor q$$</h3>

Dado que se integra la disyunción en un enunciado condicional $(p \lor q) \rightarrow r$, el diagrama de flujo será:

or.jpeg

Intentemos analizar la siguiente afirmación:

<center>
    <h4>
        <i>
"Siempre que los clientes seleccionan la "Opcion-en" o son de Filipinas, se les envia una carta de invitación."
        </i>
    </h4>  
</center>

<center>
    Podemos entonces formalizar esto en nuestra fbf como: <br>
    $p$ = El cliente ha seleccionado 'Opcion-en' <br>
    $q$ = El cliente es de Filipinas<br>
    $r$ = El sistema envía una carta de invitación<br>
      
</center>
   <h3>$$(p \lor q)\rightarrow r$$</h3>


```python
opcion_seleccionada = True
ubicacion = 'Canada'
if (opcion_seleccionada or ubicacion == 'Filipinas'):
    print('Enviada con éxito la carta de invitación.')
```

## Conjuncion

La última conectiva que abordaremos en este módulo es la conjunción o el operador "y". Las operaciones de conjunción requieren al menos dos argumentos o premisas, podemos denotarlas como $p$ y $q$ de nuevo y utilizamos la conectiva lógica $\land$.

Podríamos entonces escribir la fbf como

<h3>$$p\land q$$</h3>

Y con un diagrama de flujo como:

an.jpeg

Considere la siguiente afirmación:

<center>
    <h4>
        <i>
"Si la persona detectada por la cámara lleva máscara y lentes de sol, se activará la alarma.."
        </i>
    </h4>  
</center>

<center>
    Podemos entonces formalizar esto en la fbf como: <br>
    $p$ = La persona en la cámara lleva una máscara <br>
    $q$ = La persona de la cámara lleva lentes de sol<br>
    $r$ = Activar la alarma<br>
      
</center>
<h3>$$(p \land q)\rightarrow r$$<h3>

We can code this as:

```python
tiene_mascara = True
tiene_lentesSol = False
if (tiene_mascara and tiene_lentesSol):
    print("Activar la alarma.")
```

Notar que en la sección de implicaciones no hemos encontrado ninguna sentencia if anidada. Aplicamos un truco al respecto, que requiere discutir primero las conjunciones.

Tenemos este diagrama de flujo:

nest.jpeg

Es posible estar más familiarizado con su forma de código como:

```python
p,q = True, False
if (p):
    if (q):
        print(True)
else:
    print(False)
```

Esto puede simplificarse aún más como se muestra a continuación:

simp.jpeg

Aquí las pruebas de cómo se obtuvo:

Dada la fbf: $p \rightarrow (q\rightarrow r)$ <br>
$ = \neg p \lor (q\rightarrow r) : $ Definición de implicación<br>
$ = \neg p \lor (\neg q \lor r)  : $ Definición de implicación<br>
$ = (\neg p \lor \neg q)\lor r   : $ Regla asociativa<br>
$ = \neg(p \land q)\lor r        : $ Ley de Morgan<br>
$ = (p \land q)\rightarrow r     : $ Definición de implicación<br>
