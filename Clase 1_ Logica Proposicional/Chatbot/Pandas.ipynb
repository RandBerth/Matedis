Dado que iterrows devuelve una Serie por cada fila, no conserva los dtypes entre las filas (los dtypes se conservan entre las columnas para DataFrames).

Para preservar los dtypes mientras se itera sobre las filas, es mejor utilizar itertuples() que devuelve namedtuples de los valores y que es generalmente más rápido que iterrows.

Nunca se debe modificar algo sobre lo que se está iterando. No se garantiza que funcione en todos los casos. Dependiendo de los tipos de datos, el iterador devuelve una copia y no una vista, y escribir en él no tendrá ningún efecto.


```python
import random
lista = ["manzana", "platano", "cereza"]
print(random.choices(lista, weights = [10, 1, 1], k = 14))
```

Devuelve una lista con 14 elementos. La lista  contiene una selección aleatoria de los valores de una lista especificada, y debe haber 10 veces más posibilidades de seleccionar "manzana" que "platano" y "cereza":

El método choices() devuelve una lista con el elemento seleccionado aleatoriamente de la secuencia especificada.

Puede ponderar la posibilidad de cada resultado con el parámetro weights o el parámetro cum_weights.

La secuencia puede ser una cadena, un rango, una lista, una tupla o cualquier otro tipo de secuencia.

random.choices(sequence, weights=None, cum_weights=None, k=1)

sequence : es obligatorio, una lista, una tupla, un rango de números, etc.

weights : es opcional, lista que puede ponderar la posibilidad de cada valor, por defecto es None

cum_weights : es opcional, puede pesar la posibilidad de cada valor, esta vez se acumula.Ej: [2, 1, 1] cum_weights; [2, 3, 4], por defecto es None

k : es opcional, entero que define la longitud de la lista que devuelve.

Se importa datetime de la librería estándar de Python para obtener la hora actual.

Función para obtener el saludo adecuado:

Definimos la función obtener_saludo() que determina el saludo adecuado basado en la hora actual del sistema:

Entre 5:00 y 11:59 -> "¡Buenos días!"
Entre 12:00 y 17:59 -> "¡Buenas tardes!"
Entre 18:00 y 4:59 -> "¡Buenas noches!"

Conversión del saludo:

Llamamos a obtener_saludo() para obtener el saludo correcto y lo mostramos al iniciar el chatbot.

```python
import random
from datetime import datetime
class ChatBotFacil:
    def __init__(self, chatbote): # El metodo __init__ de instancia inicializa el objeto recién creado.
                                  # el método toma el objeto como primer argumento (self),
                                  # seguido de cualquier argumento adicional que se le deba pasar.
        self.chatbote = chatbote

    def obtener_respuesta(self, entrada_usuario):
        if entrada_usuario in self.chatbote:
            return random.choice(self.chatbote[entrada_usuario])
        else:
            return "Lo siento, no entiendo lo que dices."

def obtener_saludo():
    hora_actual = datetime.now().hour
    if 5 <= hora_actual < 12:
        return "¡Buenos días!"
    elif 12 <= hora_actual < 18:
        return "¡Buenas tardes!"
    else:
       return "¡Buenas noches!"

chat_bot = ChatBotFacil(chatbote)

print(f"Chatbot: {obtener_saludo()} ¿En qué puedo ayudarte?")#La funcion obtener saludo define la hora del dia

# Interactuar con el chatbot
while True:
    entrada_usuario = input("Usuario: ")
    if entrada_usuario.lower() in ["salir", "exit", "adiós"]:
        print("Chatbotin: ¡Adiós!")
        break
    Chatbot = chat_bot.obtener_respuesta(entrada_usuario)
    print(f"Chatbotin: {Chatbot}")
```

```python
import pandas as pd
import random
from datetime import datetime

# Cargar el archivo de Excel
df = pd.read_excel('Libro3.xlsx')

# Crear un diccionario a partir del DataFrame, normalizando a minúsculas
chatbote = {}
for index, row in df.iterrows():
    input_key = row['Usuario'].strip().lower()  # Normalizar la clave
    if input_key not in chatbote:
        chatbote[input_key] = []
    chatbote[input_key].append(row['Chatbot'])

class ChatBotFacil:
    def __init__(self, chatbote):
        self.chatbote = chatbote

    def obtener_respuesta(self, entrada_usuario):
        entrada_usuario = entrada_usuario.strip().lower()  # Normalizar la entrada del usuario
        if entrada_usuario == "¿qué hora es?":
            hora_actual = datetime.now().strftime("%H:%M")
            return f"La hora actual es {hora_actual}"
        elif entrada_usuario in self.chatbote:
            return random.choice(self.chatbote[entrada_usuario])
        else:
            return "Lo siento, no entiendo lo que dices."

def obtener_saludo():
    hora_actual = datetime.now().hour
    hora_actual = int(hora_actual)-5
    if 5 <= hora_actual < 12:
        return "¡Buenos días!"
    elif 12 <= hora_actual < 18:
        return "¡Buenas tardes!"
    else:
        return "¡Buenas noches!"

chat_bot = ChatBotFacil(chatbote)

# Mensaje de saludo basado en la hora del sistema
print(f"Chatbot: {obtener_saludo()} ¿En qué puedo ayudarte?")

# Interactuar con el chatbot
while True:
    entrada_usuario = input("Usuario: ")
    if entrada_usuario.strip().lower() in ["salir", "exit", "adiós"]:
        print("Chatbot: ¡Adiós!")
        break
    respuesta = chat_bot.obtener_respuesta(entrada_usuario)
    print(f"Chatbot: {respuesta}")
```

```python
import pandas as pd
import random
from datetime import datetime
from unidecode import unidecode
import string

# Cargar el archivo de Excel
df = pd.read_excel('Libro3.xlsx')

# Crear un diccionario a partir del DataFrame, normalizando a minúsculas, eliminando acentos y signos
chatbote = {}
def normalizar(texto):
    texto = unidecode(texto).lower().strip()
    texto = texto.translate(str.maketrans('', '', string.punctuation))
    return texto

for index, row in df.iterrows():
    input_key = normalizar(row['Usuario'])
    if input_key not in chatbote:
        chatbote[input_key] = []
    chatbote[input_key].append(row['Chatbot'])

class ChatBotFacil:
    def __init__(self, chatbote):
        self.chatbote = chatbote

    def obtener_respuesta(self, entrada_usuario):
        entrada_usuario = normalizar(entrada_usuario)  # Normalizar la entrada del usuario
        if entrada_usuario == "que hora es":
            hora_actual = datetime.now().strftime("%H:%M")
            return f"La hora actual es {hora_actual}"
        elif entrada_usuario in self.chatbote:
            return random.choice(self.chatbote[entrada_usuario])
        else:
            return "Lo siento, no entiendo lo que dices."

def obtener_saludo():
    hora_actual = datetime.now().hour
    if 5 <= hora_actual < 12:
        return "¡Buenos días!"
    elif 12 <= hora_actual < 18:
        return "¡Buenas tardes!"
    else:
        return "¡Buenas noches!"

chat_bot = ChatBotFacil(chatbote)

# Mensaje de saludo basado en la hora del sistema
print(f"Chatbot: {obtener_saludo()} ¿En qué puedo ayudarte?")

# Interactuar con el chatbot
while True:
    entrada_usuario = input("Usuario: ")
    if normalizar(entrada_usuario) in ["salir", "exit", "adios"]:
        print("Chatbot: ¡Adiós!")
        break
    respuesta = chat_bot.obtener_respuesta(entrada_usuario)
    print(f"Chatbot: {respuesta}")
```
